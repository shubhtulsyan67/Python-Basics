```python
# 3. Loops 
#1.	Write a Python program to print numbers from 1 to 10 using a loop. 

for i in range(1, 11):
    
    print(i)
```

```python
#2.	Write a Python program to find the sum of numbers from 1 to n. 

n = int(input("Enter n: "))
total = 0

for i in range(1, n + 1):
    total += i

print("Sum:", total)
```

```python
#3.	Write a Python program to print the multiplication table of a given number. 

num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
#4.	Write a Python program to calculate the factorial of a number using a loop. 

num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact *= i

print("Factorial:", fact)
```

```python
#5.	Write a Python program to check whether a given number is a prime number or not. 

num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
else:
    print("Not Prime")
```
