```python
#4. Pattern Programs 
#1.Write a Python program to print the following pattern: 

for i in range(1, 6):
    print("*" * i)
```

```python
#2. Write a Python program to print: 

for i in range(5, 0, -1):
    print("*" * i)
```

```python
#3. Write a Python program to print: 

for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

```python
#4. Write a Python program to print: 

for i in range(1, 6):
    print(str(i) * i)
```

```python
#5. Write a Python program to print: 

for i in range(1, 6):
    for j in range(i):
        print("*", end=" ")
    print()
```
